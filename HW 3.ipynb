### The Problem
Model 1 (GM ~ Toyota) has R2 = 0.41. Model 2 (GM ~ Toyota + Ford) has R2
= 0.46. Since Model 2 has higher R2, it is the better predictive model and we
should always prefer it.
### Analysis and Correction
The key error is assuming a model that adding an additional predictor cant ever decrease R^2, so Model 2 doesnt have a higher R^2 because its a better model. The additional predictor can never lower an R^2 value, even with them just being considered noise. So the moderl should instead be compared by adjusting the R^2, AIC/BIC, or out-of-sample prediction performance. This way the extra predictor can assessed to see if it does actually improve predictive performance instead of just it improving in-sample fit.
